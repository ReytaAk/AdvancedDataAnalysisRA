```python
import statsmodels.api as sm

y = data['prestg10']
X = data[['female', 'age', 'educ', 'conrinc']]

X = sm.add_constant(X)

model = sm.OLS(y, X, missing='drop')
model_result = model.fit()
model_result.summary()
```

### **Interpret the coefficients of the model, and discuss the fit and predictive capabilities**

The model seems to predict occupational prestige with a remarkably high explanatory power (R2 =.292). All of the independent variables, except age, has a statistically significant impact on the dependent variable. Although the p value of age is still around a considerable limit, since its confidence interval includes zero, it can be reported as not has a significant impact on occupational prestige. Among the others, being female, in reference to being male, significant at 95% confidence interval level (p value <= 0.05), while education and conrinc are significant at the 99% conf. interv. level (p value < 0.001). While being female results in a decrease, spesifically a -1.05 unit of decrease in occupational prestige, one unit of increase in education brings 1.8 unit of increase in prestige. Similarly, respondent's income increases occupational prestige by 8.4 unit for each unit of increase in income. (The relationship for the last one can be vice versa but anyway..)

## **6. MODEL DIAGNOSTICS**

### **Perform diagnostic tests to check the validity of the OLS model assumptions (linearity, homoscedasticity, normality, etc.)...**

```python
#Normality
import matplotlib.pyplot as plt
from scipy import stats
sns.distplot(data['prestg10'], fit=stats.norm)
```

Althoug some fluctuations, the distribution of prestg variable overall follows the nurmality curve.

```python
#Normality
skewness = stats.skew(data['prestg10'].dropna())
kurtosis = stats.kurtosis(data['prestg10'].dropna())
print(skewness, kurtosis)
```

```python
#Collinearity
sns.pairplot(X[['female', 'age', 'educ', 'conrinc']])
```

There seems to be no strong correlation among the variables in the matrix plots above. So no serious problem of multicollinearity.

```python
#Normality
import matplotlib.pyplot as plt
from scipy import stats
sns.distplot(model_result.resid, fit=stats.norm)
```

Distribution of residuals in my model is almost perfectly normally distributed as observed in the above grapgh.

```python
#Linearity and homoscedasticity
import matplotlib.pyplot as plt


fitted_values = model_result.fittedvalues  
residuals = model_result.resid  

plt.scatter(fitted_values, residuals, alpha=0.7)
plt.axhline(0, color='red', linestyle='--', linewidth=1)
plt.xlabel("Fitted Values")
plt.ylabel("Residuals")
plt.show()
```

The data points spread randomly around the red line (0 residual), shows that our model checks for both the linearity and absece of homoscedasticty assumptions.
